chi2 test is used since comparing two categorical data

```python
from scipy.stats import chi2_contingency

```

H0: Loyalty tier has a significant impact on campaign response

```python
contingency_table = pd.crosstab(df['loyalty_tier'], df['responded'])

chi2, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"P-value: {p_val}")


if p_val < 0.05:
    print("Accepting the null hypothesis: Loyalty tier has a significant impact on campaign response.")
else:
    print("Rejecting the null hypothesis: Loyalty tier has a significant impact on campaign response.")
```

H0: Engagement score has a significant impact on campaign response

```python
contingency_table = pd.crosstab(df['engagement_score'], df['responded'])

chi2, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"P-value: {p_val}")


if p_val < 0.05:
    print("Accepting the null hypothesis: Engagement score has a significant impact on campaign response.")
else:
    print("Rejecting the null hypothesis: Engagement score has a significant impact on campaign response.")
```
